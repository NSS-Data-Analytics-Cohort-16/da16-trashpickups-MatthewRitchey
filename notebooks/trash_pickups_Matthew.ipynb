```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
df_trash = pd.read_csv('../../data/trash_hauler_report_with_lat_lng.csv')
df_trash.head(10)
```

```python
# filter for only missed pick ups
missed_pickups = df_trash[df_trash['Request'].str.contains("miss", case=False, na=False)]
missed_pickups.head(10)
```

```python
# grouping by address to see how many times each location had a missed pickup
missed_counts = missed_pickups.groupby("Incident Address").size().reset_index(name="missed_count")
missed_counts.head(10)
```

```python
#applying the $200 fine per missed pickup for every missed pickup after the first
missed_counts["fine"] = (missed_counts["missed_count"] - 1).clip(lower=0) * 200
missed_counts.head(10)
```

```python
# calculate the total fines for missed pickups
total_damages = missed_counts["fine"].sum()
print("Total damages due to missed pickups: $", total_damages)
```

-----How much does each trash hauler owe?-----

```python
#breaking down the fines per trash hauler

#this query shows the amount of addresses each hauler has fines for missing pickups from
hauler_fines = missed_pickups.groupby("Trash Hauler")["Incident Address"].nunique()
hauler_fines.head(10)
```

```python
# Normalize hauler names to uppercase
missed_pickups.loc[:, "Trash Hauler"] = missed_pickups["Trash Hauler"].str.upper()

# Regroup by hauler and count unique addresses
hauler_fines = missed_pickups.groupby("Trash Hauler")["Incident Address"].nunique()
print(hauler_fines.head(10))
```

```python
# Count missed pickups per address + hauler
missed_counts = (
    missed_pickups
    .groupby(["Incident Address", "Trash Hauler"])
    .size()
    .reset_index(name="missed_count")
)

# Apply fine rule
missed_counts["fine"] = (missed_counts["missed_count"] - 1).clip(lower=0) * 200

# Normalize hauler names
missed_counts["Trash Hauler"] = missed_counts["Trash Hauler"].str.upper()

# Aggregate by hauler
hauler_summary = (
    missed_counts
    .groupby("Trash Hauler")
    .agg(
        total_missed=("missed_count", "sum"),
        total_fines=("fine", "sum")
    )
    .reset_index()
)

print(hauler_summary.head(10))
```

-----What were the total missed pickup by route?-----

```python
# ‚¶Å	What were to total missed pickup by route?
missed_counts_route = (
    missed_pickups
    .groupby(["Incident Address", "Trash Route"])
    .size()
    .reset_index(name="missed_count")
)

# Apply fine rule
missed_counts_route["fine"] = (missed_counts_route["missed_count"] - 1).clip(lower=0) * 200

# Aggregate by route
route_summary = (
    missed_counts_route
    .groupby("Trash Route")
    .agg(
        total_missed=("missed_count", "sum"),
        total_fines=("fine", "sum")
    )
    .reset_index()
)

# View results
print(route_summary.head(10))
```

```python
# Count missed pickups per address + district
missed_counts_district = (
    missed_pickups
    .groupby(["Incident Address", "Council District"])
    .size()
    .reset_index(name="missed_count")
)

# Apply fine rule
missed_counts_district["fine"] = (
    (missed_counts_district["missed_count"] - 1).clip(lower=0) * 200
)

# Aggregate by district
district_summary = (
    missed_counts_district
    .groupby("Council District")
    .agg(
        total_missed=("missed_count", "sum"),
        total_fines=("fine", "sum")
    )
    .reset_index()
)

# View results
print(district_summary.head(50))
```

```python
# Alternative Method 1:
# For each address, if there are three or more missed pickups within a 180-day period, damages of $1500 
# will be charged. (A fine will be levied every time three unique missed pickup dates occur within a six
# -month period for a single address.)

alt_1_df = df_trash.copy()
alt_1_df['Date Opened'] =pd.to_datetime(alt_1_df['Date Opened'])
alt_1_df = alt_1_df.sort_values(['Incident Address', 'Date Opened'])

# for each pickup, count how many pickups happened in the last 180 days (including itself)
rolling_count = alt_1_df.groupby('Incident Address').rolling(window='180d', on='Date Opened')['Date Opened'].count()

# charge $1,500 every time the count reaches 3 or more
total_fines_method1 = (rolling_count >= 3).sum() * 1500

print(f"Alternative Method 1 Total Fines: ${total_fines_method1:,.0f}")
```

```python
# Alternative Method 2:
# This method also considers the six-month window like Alternative Method 1, but each date can only be
# used once to support a fine. How will this difference impact the fines levied? (Example: If Jan 1st,
# Mar 3rd, Apr 8th, and Aug 9th were the only four dates a trash pickup was missed, the original method 
# would result in $3000 in fines [Jan, Mar, Apr, and also Mar, Apr, Aug]. However, this updated method
# would only result in $1500 because neither Mar nor Apr can be used for another fine since they were 
# already used.)
alt_2 = df_trash.copy()
alt_2['Date Opened'] =pd.to_datetime(alt_2['Date Opened'])
alt_2 = alt_2.sort_values(['Incident Address', 'Date Opened'])

total_fines_method2 = 0

# process each address
for address in alt_2['Incident Address'].unique():
    address_dates = alt_2[alt_2['Incident Address'] == address]['Date Opened'].tolist()
    used_dates = set()

    #go through dates abd find set of 3 within 180 days
    for i in range(len(address_dates)):
        if address_dates[i] in used_dates:
            continue

        # find dates within 180 days that haven't been used
        window = [address_dates[i]]
        for j in range(i+1, len(address_dates)):
            if address_dates[j] not in used_dates:
                if (address_dates[j] - address_dates[i]).days <= 180:
                    window.append(address_dates[j])

        # if we have 3+ dates charge $1500 and mark them as used
        if len(window) >= 3:
            total_fines_method2 += 1500
            used_dates.update(window[:3]) # use only the first 3 dates

print(f"Alternative Method 2 Total Fines: ${total_fines_method2:,.0f}")
```
